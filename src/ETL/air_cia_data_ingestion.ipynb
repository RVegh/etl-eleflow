```python
import pandas as pd
import csv
import os
import glob
```

```python
pd.set_option('display.max_columns', None)
```

```python
#Capturando todos os csvs do path abaixo e concatenando para criar um único dataset.
file_path = r'C:\Users\Vegh\Documents\Estudo\data_engineer_test\source_files\AIR_CIA'
database_path = r'C:\Users\Vegh\Documents\Estudo\data_engineer_test\database_files'

files = glob.glob(os.path.join(file_path, '*.csv'))
df_list = []

for file in files:
    df = pd.read_csv(file, sep=';', index_col=0, header=0)
    df_list.append(df)

df = pd.concat(df_list, axis=0, ignore_index=True)
    
    
```

```python
df.head()
```

```python
#Renomeando colunas para snake_case.
df.columns = ['ICAO_IATA', 'CNPJ', 'ATIVIDADES_AEREAS', 
              'ENDERECO_SEDE', 'TELEFONE', 'EMAIL', 
              'DECISAO_OPERACIONAL', 'DATA_DECISAO_OPERACIONAL', 
              'VALIDADE_OPERACIONAL']
df.sample()
```

```python
#Criando duas novas colunas a partir da coluna ICAO_IATA
#df.drop(df['ICAO_IATA]) Caso fosse necessário deletar a coluna original

df[['ICAO', 'IATA']] = df['ICAO_IATA'].str.split(' ', n=1, expand=True).fillna('Nulo')
```

```python
#Substituindo os espaços em branco para criar códigos únicos(NÃO OBRIGATÓRIO)

df['ICAO_IATA'] = df['ICAO_IATA'].str.replace(' ', '')
df.sample()
```

```python
#Save tranformed data into csv
df.to_csv( os.path.join(database_path, 'air_cia.csv'), sep=';', index=False)
```
